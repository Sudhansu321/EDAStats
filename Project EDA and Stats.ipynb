```python
# 10. Check the distribution of columns. List down columns that are normally
    ## distributed. List down columns that are not normally distributed.
```

```python
from scipy.stats import skew, kurtosis

# Calculating  skewness and kurtosis for each column

distribution_info = []
for column in df.columns:
    column_stats = {
        'Column': column,
        'Skewness': skew(df[column]),
        'Kurtosis': kurtosis(df[column])
    }
    distribution_info.append(column_stats)

# Separate columns into normally and non-normally distributed based on skewness and kurtosis
normally_distributed = []
non_normally_distributed = []
for column_stats in distribution_info:
    column = column_stats['Column']
    skewness = column_stats['Skewness']
    kurtosis_val = column_stats['Kurtosis']
    
    if abs(skewness) <= 0.5 and abs(kurtosis_val) <= 1.0:
        normally_distributed.append(column)
    else:
        non_normally_distributed.append(column)

print("Columns that are normally distributed:")
print(normally_distributed)
print()

print("Columns that are not normally distributed:")
print(non_normally_distributed)
```

```python
# Calculating  skewness for each column

skewness = df.skew()

# Set the threshold for identifying highly skewed columns

skewness_threshold = 1.0

# Identify highly skewed columns

highly_skewed_columns = skewness[skewness.abs() > skewness_threshold].index.tolist()


print("Columns that are highly skewed:")
print(highly_skewed_columns)
```

```python
# 15. Find the distribution of all variables with respect to the outcome ‘Class’ variable.
```

```python
# Selecting all variables except 'Class'

variables = df.columns.drop('Class')

# Loop through each variable and plot distribution with respect to 'Class'

for var in variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=var, hue='Class', kde=True, bins=50)
    plt.title(f'Distribution of {var} with respect to Class')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.legend(title='Class', labels=['Non-Fraud', 'Fraud'])
    plt.show()
```
